<h1>Data Cleaning</h1>

```python
import pandas as pd
import numpy as np
import nltk
import contractions 
from nltk.tokenize import word_tokenize 
from nltk.corpus import stopwords, wordnet
from nltk.stem import WordNetLemmatizer 
import re
```

```python
df = pd.read_csv('./CSV/WheyData.csv')
```

```python
df.head()
```

```python
type(df['text'][0])
```

```python
df = df.drop(['name'], axis = 1)
```

```python
df = df.fillna('0')
```

```python
mydf = df.loc[:,['rating','text','helpful_vote']]
```

```python
mydf['no_contract'] = mydf['text'].apply(lambda x: [contractions.fix(word) for word in x.split()])
```

```python
mydf.head()
```

```python
mydf['review'] = [' '.join(map(str,l)) for l in mydf['no_contract']]
```

```python
mydf.head()
```

```python
mydf = mydf.drop(['no_contract'], axis = 1)
```

```python
mydf.head()
```

```python
mydf['review'] = mydf['review'].apply(lambda x: re.sub('[^A-Za-z]',' ', x))
```

```python
type(mydf['review'][0])
```

```python
mydf['review'] = mydf['review'].apply(lambda x: x.lower())
```

```python
mydf.head()
```

```python
mydf['tokens'] = mydf['review'].apply(word_tokenize)
```

```python
mydf.head()
```

```python
stop_words = set(stopwords.words('english'))
mydf['stopwords_removed'] = mydf['tokens'].apply(lambda x: [word for word in x if word not in stop_words])
```

```python
mydf.head()
```

```python
mydf['lemmatized'] = mydf['stopwords_removed'].apply(lambda x: [WordNetLemmatizer().lemmatize(word) for word in x])
```

```python
mydf.head()
```

<h1> Exploratory Data Analysis</h1>

```python
from wordcloud import WordCloud, ImageColorGenerator
import matplotlib.pyplot as plt
from nltk.probability import FreqDist
import seaborn as sns
```

```python
words = mydf['lemmatized']
allwords = []

for wordlist in words:
    allwords += wordlist
```

```python
type(allwords)
```

<h2>Word Cloud</h2>

```python
mostcommon = FreqDist(allwords).most_common(100)

wordcloud = WordCloud(width=1000,height=600,background_color='white').generate(str(mostcommon))
fig = plt.figure(figsize=(20,6), facecolor='white')
plt.imshow(wordcloud,interpolation='bilinear')
plt.axis('off')
plt.title('Top 100 Most Common Words', fontsize=50)
plt.tight_layout(pad=0)
plt.savefig('Wordcloud.png')
```

<h2>Frequency of Most Common Words</h2>

```python
mostcommon_small = FreqDist(allwords).most_common(25)
x, y = zip(*mostcommon_small)

plt.figure(figsize=(30,20))
plt.margins(0.02)
plt.bar(x,y)
plt.xlabel('Words',fontsize=40)
plt.ylabel('Frequency of Words', fontsize=40)
plt.yticks(fontsize=15)
plt.xticks(rotation=60, fontsize=15)
plt.title('Frequency of 25 Most Common Words', fontsize=60)
plt.savefig('Frequency Count.png')
```

```python
mydf['lemma_str'] = [' '.join(map(str,l)) for l in mydf['lemmatized']]
mydf['word_count'] = mydf['lemmatized'].apply(lambda x: len(str(x).split()))
mydf['review_len'] = mydf['lemma_str'].astype(str).apply(len)
```

```python
mydf.head()
```

```python
mydf['helpful_vote'] = mydf['helpful_vote'].replace('One person found this helpful', '1')
mydf.loc[:, ['rating', 'helpful_vote']] = mydf.loc[:, ['rating', 'helpful_vote']].astype(int)
```

```python
type(mydf['helpful_vote'])
```

```python
x = mydf['helpful_vote'].tolist()
y = mydf['review_len'].tolist()

# sorted_x = [x for _,x in sorted(zip(y,x))]
```

```python
plt.figure(figsize=(30,20))
plt.margins(0.02)
plt.scatter(x,y)
plt.xlim([-1,50])
plt.xticks(fontsize = 20)
plt.yticks(fontsize = 20)
plt.title('Review Length vs Helpful',fontsize = 25)
plt.xlabel('Helpful Votes',fontsize = 20)
plt.ylabel('Review Length', fontsize = 20)
plt.savefig('Review len vs Helpfulness')
```

```python
total_helpful_votes = dict([])
for i in range(mydf.shape[0]) :
    for word in set(mydf.lemmatized[i]) :
        total_helpful_votes[word] = total_helpful_votes.get(word, 0) + mydf.helpful_vote[i]
```

```python
word_impact = pd.Series(total_helpful_votes) / mydf.helpful_vote.sum() * 100
```

```python
word_impact = word_impact.sort_values(ascending = False)
# word_impact = word_impact[word_impact > 1]
plt.figure(figsize=(30,20))
plt.bar(word_impact.index[:40],word_impact.values[:40])
plt.xticks(rotation=60,fontsize = 15)
plt.yticks(fontsize = 15)
plt.xlabel('Words',fontsize=20)
plt.ylabel('Percentage of Helpfulness', fontsize=20)
plt.title('Word Helpfulness',fontsize = 30)
plt.savefig('impact.png')
```
