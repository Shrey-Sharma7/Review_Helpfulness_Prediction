<h3>VADER</h3>

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
vds = SentimentIntensityAnalyzer()
```

```python
text = "I went to the movie,yesterday. It was amazing! Everyone acted well"
vds.polarity_scores(text)
```

<h4>Vader Sentiment Score</h4>
Positive sentiment: compound score >= 0.05 

Neutral sentiment: (compound score > -0.05) and (compound score < 0.05)

Negative sentiment: compound score <= -0.05


```python
mydf['scores'] = mydf['review_text'].apply(lambda review:vds.polarity_scores(review))
```

```python
mydf.head()
# df.head()
```

```python
mydf['compound'] = mydf['scores'].apply(lambda x:x['compound'])
```

```python
mydf.head()
```

```python
mydf['vader_score'] = mydf['compound'].apply(lambda x:'pos' if x >= 0  else 'neg')
```

```python
mydf.head()
```

```python
mydf['vader_score'].value_counts()
```

```python
mydf = mydf.drop(['scores','compound'],axis = 1)
```

Average Word Length

```python
def avg_word(sentence):
    words = sentence.split()
    return (sum(len(word) for word in words)/len(words))

mydf['avg_word_length'] = mydf['review_text'].apply(lambda x:avg_word(x))
```

```python
mydf.head()
```

No. of Stopwords

```python
stop = stopwords.words('english')
mydf['stopwords_count'] = mydf['review_text'].apply(lambda x:len([x for x in x.split() if x in stop]))
```
